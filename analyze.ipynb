```python
import pandas as pd

def func(category):
    # Load your CSV
    df = pd.read_csv(f"equiv/{category}_equivalencies.csv")

    df = df.loc[df["Status"] != "Not Equivalent"]

    df_unique = df.drop_duplicates(subset=["External Institution", "McGill Course"])

    # Count how many equivalencies each university has
    counts = df_unique["External Institution"].value_counts()

    # Show the top ones
    print("Universities with the most McGill course equivalencies:")
    print(counts.head())

    # Build pivot table
    pivot = pd.crosstab(df["External Institution"], df["McGill Course"])

    pivot = pivot.map(lambda x: x if x > 0 else "")
    pivot["Total Distinct Courses"] = counts

    pivot = pivot.sort_values("Total Distinct Courses", ascending=False)

    pivot.to_csv(f"grids/{category}_course_grid.csv")

[func(category) for category in ["CORE", "IMPACT", "MINOR", "TCA", "TCB"]]
```
